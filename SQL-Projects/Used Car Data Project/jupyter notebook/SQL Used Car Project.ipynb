```python
# Connecting to the mySQL server
import os
import mysql.connector
import pandas as pd
import matplotlib.pylab as plt
from dotenv import load_dotenv
load_dotenv()
host = os.getenv("MYSQL_HOST")
user = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PASSWORD")
database = os.getenv("MYSQL_DB")

db = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
```

1. What are the top 5 most expensive cars in the dataset based on price?

```python
Query1 = pd.read_sql_query("""
    SELECT ROW_NUMBER() OVER (ORDER BY avg(Price) DESC) AS Number, Brand, Model, AVG(Price) AS Average
    FROM `car_price_dataset`
    GROUP BY Brand, Model
    ORDER BY Average DESC
    LIMIT 5;
""", db)
Query1
```

2. What is the average price of cars grouped by year of manufacture?

```python
Query2 = pd.read_sql_query("""
    SELECT Brand, Model, COUNT(Model) AS '# of Cars', Year, AVG(Price) AS 'Average Price'
    FROM car_price_dataset
    GROUP BY Brand , Model , Year
    ORDER BY Brand ASC , Model ASC , Year DESC
""", db)
Query2
```

3. Which year has the highest average price for cars?

```python
Query3 = pd.read_sql_query("""
    SELECT ROW_NUMBER() OVER (ORDER BY avg(Price) DESC) AS Number, Year, ROUND(AVG(price), 2) as 'Average Price'
    FROM car_price_dataset
    GROUP BY Year
    ORDER BY 'Average Price' DESC
""", db)
Query3
```

4. Which brand has the highest average price for cars with automatic transmission?

```python
Query4 = pd.read_sql_query("""
    SELECT ROW_NUMBER() OVER (ORDER BY avg(Price) DESC) AS Number, Brand, Transmission, ROUND(AVG(Price), 2) as 'Average Price'
    FROM car_price_dataset
    WHERE Transmission = 'Automatic'
    GROUP BY Brand, Transmission
    ORDER BY AVG(Price) DESC;
""", db)
Query4
```

5. Which manufacturer has the most cars listed in the dataset?

```python
Query5 = pd.read_sql_query("""
    SELECT Brand, COUNT(model) AS 'Number of Cars'
    FROM car_price_dataset
    GROUP BY Brand
    ORDER BY COUNT(model) DESC;
""", db)
Query5
```

6. Which fuel type (e.g., gasoline, diesel, electric) is most common in the dataset?

```python
Query6 = pd.read_sql_query("""
    SELECT Fuel_Type, COUNT(Fuel_Type) AS '# of Cars'
    FROM car_price_dataset
    GROUP BY Fuel_Type
    ORDER BY COUNT(Fuel_Type) DESC
""", db)
Query6
```

7. Which manufacturer produces the highest number of electric vehicles in the dataset?

```python
Query7 = pd.read_sql_query("""
    SELECT Brand, COUNT(Fuel_Type) AS '# of Electric Cars'
    FROM car_price_dataset
    WHERE Fuel_Type = 'electric'
    GROUP BY Brand
    ORDER BY COUNT(Fuel_Type) DESC;
""", db)
Query7
```

8. What is the distribution of cars based on body type (e.g., sedan, SUV, coupe, hatchback)?

```python
Query8 = pd.read_sql_query("""
    SELECT 
        Brand,
        CASE
            WHEN Doors = 2 THEN 'Coupe'
            WHEN Doors = 3 THEN 'Hatchback'
            WHEN Doors = 4 THEN 'Sedan'
            WHEN Doors = 5 THEN 'SUV'
        END AS Car_Type,
        COUNT(Doors) AS '# of Cars'
    FROM car_price_dataset
    GROUP BY Brand , Doors
    ORDER BY Brand ASC;
""", db)
Query8
```

9. How have used electric vehicle prices changed over the years?

```python
Query9 = pd.read_sql_query("""
    SELECT Fuel_Type, Year, AVG(Price) AS 'Average Price'
    FROM car_price_dataset
    WHERE Fuel_Type = 'Electric'
    GROUP BY Fuel_Type , Year
    ORDER BY AVG(Price) DESC
""", db)
Query9
```

10. Which car has the lowest mileage but has had more than two owners?

```python
Query10 = pd.read_sql_query("""
    SELECT Distinct Brand, Model, Year, MIN(Mileage) AS 'Milage', Owner_Count
    FROM car_price_dataset
    GROUP BY Brand , Model, Year, Owner_Count
    HAVING Owner_Count >= 2
    ORDER BY Min(Mileage) ASC
""", db)
Query10
```
